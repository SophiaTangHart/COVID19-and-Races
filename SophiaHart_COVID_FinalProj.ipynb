```python
import pandas as pd
import numpy as np
```

```python
# Data for COVID 
COVIDdata = pd.read_csv("COVIDdata_Racial.csv")
    
COVIDdata.shape
COVIDdata.head()
```

```python
extractData = COVIDdata[["Date", "State", "Cases_Total", 'Cases_White', 'Cases_Black', 'Cases_Asian', 
                         'Cases_AIAN', 'Cases_Ethnicity_Hispanic',
                        "Deaths_Total", 'Deaths_White', 'Deaths_Black', 'Deaths_Asian', 
                         'Deaths_AIAN', 'Deaths_Ethnicity_Hispanic']]

# extract most recent date for 56 states
extractData = extractData.iloc[0:56]
print(extractData.shape)

extractData
```

```python
# remove NAN

print(extractData.isnull().values.any())

cleanCOVIDdata = extractData.dropna()
print(cleanCOVIDdata.shape)

cleanCOVIDdata
```

```python
# Data for population for different ethnic group for each state

popData = pd.read_csv('populationbyRace2019.csv')
popData
```

```python
# total number of states in COVID data, some states are removed because of NAN
n_states = len(cleanCOVIDdata.iloc[:,1])
print(n_states)

#total states in population file
p_states = len(popData.iloc[:,0])
print(p_states)
     
R_cases_total = []
R_cases_white = []
R_cases_black = []
R_cases_hispanic = []
R_cases_asian = []
R_cases_indian = []

R_deaths_total = []
R_deaths_white = []
R_deaths_black = []
R_deaths_hispanic = []
R_deaths_asian = []
R_deaths_indian = []

# store ratios for each state
Ratios = pd.DataFrame()


for s in range(n_states):
    state_in_COVID = cleanCOVIDdata.iloc[s,]
   
    for p in range(p_states):
        state_in_pop = popData.iloc[p,]
        
        if state_in_COVID['State'] == state_in_pop['State']:
            r_t = state_in_COVID['Cases_Total'] / state_in_pop['Total']
            r_t = round(r_t, 3)
            R_cases_total.append(r_t)
            
            r_w = state_in_COVID['Cases_White'] / state_in_pop['White']
            r_w = round(r_w, 3)
            R_cases_white.append(r_w)
            
            r_b = state_in_COVID['Cases_Black'] / state_in_pop['Black']
            r_b = round(r_b, 3)
            R_cases_black.append(r_b)
            
            r_h = state_in_COVID['Cases_Ethnicity_Hispanic'] / state_in_pop['Hispanic']
            r_h = round(r_h, 3)
            R_cases_hispanic.append(r_h)
            
            r_a = state_in_COVID['Cases_Asian'] / state_in_pop['Asian']
            r_a = round(r_a, 3)
            R_cases_asian.append(r_a)
            
            r_i = state_in_COVID['Cases_AIAN'] / state_in_pop['Indian']
            r_i = round(r_i, 3)
            R_cases_indian.append(r_i)
            
            
            r_d_t = state_in_COVID['Deaths_Total'] / state_in_pop['Total']
            r_d_t = round(r_d_t, 5)
            R_deaths_total.append(r_d_t)
            
            r_d_w = state_in_COVID['Deaths_White'] / state_in_pop['White']
            r_d_w = round(r_d_w, 5)
            R_deaths_white.append(r_d_w)
            
            r_d_b = state_in_COVID['Deaths_Black'] / state_in_pop['Black']
            r_d_b = round(r_d_b, 5)
            R_deaths_black.append(r_d_b)
            
            r_d_h = state_in_COVID['Deaths_Ethnicity_Hispanic'] / state_in_pop['Hispanic']
            r_d_h = round(r_d_h, 5)
            R_deaths_hispanic.append(r_d_h)
            
            r_d_a = state_in_COVID['Deaths_Asian'] / state_in_pop['Asian']
            r_d_a = round(r_d_a, 5)
            R_deaths_asian.append(r_d_a)
            
            r_d_i = state_in_COVID['Deaths_AIAN'] / state_in_pop['Indian']
            r_d_i = round(r_d_i, 5)
            R_deaths_indian.append(r_d_i)
                    
            continue

            
Ratios['State'] = cleanCOVIDdata.iloc[:, 1]
Ratios['case_Total'] = R_cases_total
Ratios['case_White'] = R_cases_white
Ratios['case_Black'] = R_cases_black
Ratios['case_Hispanic'] = R_cases_hispanic
Ratios['case_Asian'] = R_cases_asian
Ratios['case_Indian'] = R_cases_indian

Ratios['death_Total'] = R_deaths_total
Ratios['death_White'] = R_deaths_white
Ratios['death_Black'] = R_deaths_black
Ratios['death_Hispanic'] = R_deaths_hispanic
Ratios['death_Asian'] = R_deaths_asian
Ratios['death_Indian'] = R_deaths_indian
Ratios
```

```python
# expert dataframe to csv file

Ratios.to_csv('COVID_CasesDeathsRatios_byRaceState.csv')
```
